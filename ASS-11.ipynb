```python
import pandas as pd
import matplotlib.pyplot as plt

# Simulate the data if 'population.csv' is not found
data = {
    "Country": ["China", "India", "United States", "Indonesia", "Pakistan"],
    "Population": [1412000000, 1393000000, 332000000, 276000000, 225000000]
}
df = pd.DataFrame(data)

# Original line causing the error, commented out:
# df = pd.read_csv("population.csv")
df = df.sort_values("Population", ascending=False).head(5)

plt.bar(df["Country"], df["Population"])
plt.ylabel("Population")
plt.title("Top 5 Countries by Population")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Simulate the data if 'movies_genre.csv' is not found
data = {
    "Genre": ["Action", "Comedy", "Drama", "Sci-Fi", "Horror"],
    "Movie_Count": [120, 150, 180, 90, 70]
}
df = pd.DataFrame(data)

plt.bar(df["Genre"], df["Movie_Count"])
plt.xticks(rotation=45)
plt.ylabel("Movie Count")
plt.title("Movies Produced by Genre")
plt.show()
```
